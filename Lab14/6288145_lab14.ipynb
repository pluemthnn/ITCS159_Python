<h2>Exercise 11</h2>

```python
import numpy as np
test_image = img.copy()
lx , ly , lz = test_image.shape
X, Y = np. ogrid[0:lx , 0: ly]
mask = (X - lx / 2) ** 2 + (Y - ly / 2) ** 2 > lx * ly / 4; # change this to switch shape of masking crop
test_image [ mask ]=0;
io.imshow ( test_image )
io.show()
```

<h2>Exercise 13</h2>

```python
from skimage.transform import resize , rotate , swirl 
resized = resize(img , (350 , 300) , mode ='constant',anti_aliasing = True)
io.imshow( resized )
io.show()
```

```python
resized = resize(img , (50 , 50) , mode ='constant',anti_aliasing = True)
io.imshow( resized )
io.show()
```

<h2>Exercise 14</h2>

```python
# Rotate image by a certain angle around its center .
# Rotation angle in degrees in counter - clockwise direction .
rotate = rotate(img , 180)
io.imshow( rotate )
io.show()
```

<h2>Exercise 15</h2>

```python
swirled = swirl(img , rotation =0 , strength =100 , radius =120 ,mode ='constant')
io.imshow( swirled )
io.show()
```

<h2>Exercise 16</h2>

```python
from skimage import filters
blurred_img = filters . gaussian (img , sigma =3 , multichannel =True )
io. imshow ( blurred_img )
io. show ()
```

```python
blurred_img = filters . gaussian (img , sigma =5 , multichannel =True )
io. imshow ( blurred_img )
io. show ()
# change sigma to blur more
```

<h2>Exercise 17</h2>

```python
alpha = 0.15
shapened_img = img + alpha * ( img - blurred_img )
io.imshow( shapened_img . astype ( int) )
io.show()
```
